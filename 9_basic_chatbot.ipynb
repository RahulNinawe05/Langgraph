```python
from langgraph.graph import StateGraph , START, END
from typing import TypedDict,Annotated
from langchain_groq import ChatGroq
from dotenv import load_dotenv
from langchain_core.messages import BaseMessage, HumanMessage
import os
from langgraph.checkpoint.memory import InMemorySaver
from langchain_core.runnables.graph import MermaidDrawMethod
```

```python
api_key = os.getenv("GROQ_API_KEY")
llm = ChatGroq(groq_api_key=api_key, model="llama-3.1-8b-instant")
```

```python
from langgraph.graph.message import add_messages
class classState(TypedDict):
    
    messages: str
```

```python
def Chat_Node(state: classState):

    messages = state['messages']

    response = llm.invoke(messages)

    return {'messages':[response]}
```

```python
graph = StateGraph(classState)

graph.add_node('Chat_Node', Chat_Node)

graph.add_edge(START, 'Chat_Node')
graph.add_edge('Chat_Node', END)

checkpointer = InMemorySaver()
chatbot = graph.compile()
print(chatbot)
```

```python
initial_state = {
    "messages": [HumanMessage(content="who is the Rahul Gandi")]
}

chatbot.invoke(initial_state)['messages'][-1].content
```

```python
thread_id = '1'

while True:

    user_input = input("Type Here,You can exit say ['exit','quit' or 'bye']")

    print('User', user_input)

    if user_input.strip().lower() in ['exit','quit', 'bye']:
        break
    
    config = {'configurable': {'thread_id': thread_id}}
    response = chatbot.invoke({'messages':[HumanMessage(content=user_input)]}, config=config)

    print('AI: ', response['messages'][-1].content)
```
