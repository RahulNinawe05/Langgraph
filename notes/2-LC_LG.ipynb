# Workflow vs Agent

## Workflow
A **workflow** is a predefined sequence of steps —  
like a **pipeline** where data or tasks move in a fixed order.

### Example: Company “Leave Approval Workflow”
1. Employee applies for leave  
2. Manager approves or rejects  
3. HR updates the system  

Everything happens in a **fixed order**.  
If any step fails, the workflow **stops or rolls back**.

---

## Agent
An **Agent** can **think**, **plan**, and **decide** based on given instructions —  
it’s **not fixed** like a workflow.  
It dynamically chooses what to do next depending on the situation.

### Example: AI Agent Task
> “Get me the latest AI trends and make a summary.”

### What it does:
1. Searches the web  
2. Reads multiple sources  
3. Writes a summary  
4. If needed, generates an image or graph  

The **Agent decides each step on its own** —  
you don’t need to guide it manually every time.


#  Challenges & LangGraph Overview

## Challenges in Traditional Workflows
- **Condition Branches:**  
  Logic-based decisions (if/else paths).

- **Loops:**  
  Repeating steps until a goal or condition is met.

- **Jumps:**  
  Skipping between non-sequential steps as per logic.

---

## The Problem
In many LLM pipelines or LangChain workflows,  
these structures (branches, loops, jumps) require **manual “glue code”** —  
usually written in **Python**, **not** handled directly by LangChain.

---
## Why Use Langgraph
- LangChain → Linear chains / basic agents
- LangGraph → Non Linear chains , Smart, branching, looping, event-based agents, solve glue code.

## What does “State” mean?
- system ya agent ke paas abhi tak jo data, memory, ya progress hai.  
Yaani, system ne ab tak kya kiya, kya output mila, user ne kya bola —  
ye sab mila ke “state” banti hai.

- state connect rahate he all nodes ko



# Fault tolerance —
system error hone par bhi crash na ho, balki recover ho jaaye aur kaam continue kare.
- Yaani:
    - Agar ek step me kuch galti ho gayi 😬 —
    - toh pura system band nahi hota,
    - balki intelligently next best action leta hai (retry, skip, fallback, etc.).

# Event-driven system
ek aisa system jo events (signals, triggers) ke base par kaam karta hai,
na ki fixed sequence (like linear workflow) par.
- Yaani —
    - jab koi “event” hota hai (user message, API response, tool ka output, error, etc.),
    - tab system automatically decide karta hai ab kya karna hai next.
- **langchain is not even driven but solved langgraph**


# Nested workflow
- ek workflow ke andar doosra workflow.

- Yaani ek main process hai jisme sub-processes (chhote workflows) call hote hain.
Jaise programming me “function ke andar function” hota hai,
waise hi automation me “workflow ke andar workflow”.

- 🔹 Simple Example (Real Life):
    Socho tum ek Online Order System bana rahe ho 
    Main workflow:
    Take customer order
    Process payment
    Deliver item
    Ab “Process payment” ke andar bhi ek chhota workflow ho sakta hai:
    Verify card → Deduct money → Send payment receipt

- Ye hi nested workflow hai.
